# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby("type").count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby("type").sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fare = pyber_data_df.groupby("type").sum()["fare"]
total_fare
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fare / total_rides
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fare / total_drivers
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides" : total_rides,
    "Total Drivers" : total_drivers,
    "Total Fares" : total_fare,
    "Average Fare per Ride" : avg_fare_per_ride,
    "Average Fare per Driver" : avg_fare_per_driver,
    "No of Drivers per Ride": total_drivers / total_rides
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_formatted_df = pd.DataFrame({
    "Total Rides" : pyber_summary_df["Total Rides"],
    "Total Drivers" : pyber_summary_df["Total Drivers"],
    "Total Fares" : pyber_summary_df["Total Fares"].map("${:,.2f}".format),
    "Average Fare per Ride" : pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format),
    "Average Fare per Driver" : pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format),
    "No of Drivers per Ride": pyber_summary_df["No of Drivers per Ride"].map("{:,.2f}".format)
})
pyber_summary_formatted_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fare_by_date = pyber_data_df.groupby(by=["type","date"]).sum()[["fare"]]
total_fare_by_date
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fare_by_date = total_fare_by_date.reset_index()
total_fare_by_date
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fare_by_date_pivot = total_fare_by_date.pivot(index='date',columns='type',values='fare')
total_fare_by_date_pivot
```

```python
total_fare_by_date_pivot.info()
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
total_fare_by_date_pivot_loc = total_fare_by_date_pivot.loc['2019-01-01':'2019-04-28'] 
total_fare_by_date_pivot_loc
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
total_fare_by_date_pivot_loc.index = pd.to_datetime(total_fare_by_date_pivot_loc.index)
total_fare_by_date_pivot_loc
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
total_fare_by_date_pivot_loc.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
total_fare_by_week = total_fare_by_date_pivot_loc.resample("W").sum()
total_fare_by_week
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.dates as mdates
months = mdates.MonthLocator()  # every month
months_fmt = mdates.DateFormatter('%b')

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(20,6))
ax.plot(total_fare_by_week['Rural'],label='Rural')
ax.plot(total_fare_by_week['Suburban'],label='Suburban')
ax.plot(total_fare_by_week['Urban'],label='Urban')


ax.set_xlabel("Month")
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)

ax.set_ylabel("Fare ($USD)")

plt.title("Total Fare by City Type")
ax.legend()

plt.savefig("analysis\PyBer_fare_summary.png")
plt.show()
```

## EXTRA: Analyze the weekly number of rides per City Type

```python
# E9. Using groupby() on the pyber_data_df to create a new DataFrame showing the total number of rides
#  for each date where the indices are the city type and date.
total_rides_by_date = pyber_data_df.groupby(by=["type","date"]).count()[["ride_id"]]
total_rides_by_date
```

```python
# E10 - Flatten the table to pivot
total_rides_by_date = total_rides_by_date.reset_index()
total_rides_by_date
```

```python
#E11 - Create Pivot table
total_rides_by_date_pivot = total_rides_by_date.pivot(index='date',columns='type',values='ride_id')
total_rides_by_date_pivot
```

```python
#E12 Extract data from 01-01-2019 to 04-28-2019
total_ride_by_date_pivot_loc = total_rides_by_date_pivot.loc['2019-01-01':'2019-04-28']
total_ride_by_date_pivot_loc
```

```python
#E13 - set index as date time
total_ride_by_date_pivot_loc.index = pd.to_datetime(total_ride_by_date_pivot_loc.index)
total_ride_by_date_pivot_loc
```

```python
#E14 Resample the date by Week
total_ride_by_week = total_ride_by_date_pivot_loc.resample("W").sum()
total_ride_by_week
```

```python
#E15 Create Graph

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(20,6))
ax.plot(total_ride_by_week['Rural'],label='Rural')
ax.plot(total_ride_by_week['Suburban'],label='Suburban')
ax.plot(total_ride_by_week['Urban'],label='Urban')


ax.set_xlabel("Month")
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(months_fmt)

ax.set_ylabel("Rides")

plt.title("Total Rides by City Type")
ax.legend()

plt.savefig("analysis\PyBer_ride_summary.png")
plt.show()
```
