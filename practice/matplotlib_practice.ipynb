```python
#import dependecies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the line plot
plt.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plot()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#create plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis,y_axis,label='Boston', color="blue",marker="*",linewidth=2)

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# set the x limit between jan and june
#plt.xlim("Jan","June")
plt.xlim(0, 5)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()

# Add figure size
plt.figure(figsize=(20,10))
```

```python
# Create chart using OO functions
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x_axis,y_axis,label='Boston',color="green",marker="D",linewidth=2)

ax.set_ylim(0,45)
ax.set_xlim("Jan","Dec")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

ax.set_title("PyBer Dare by Month")

ax.legend()

ax.grid()
```

```python
# Create the bar plot
plt.bar(x_axis,y_axis,color="green",label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# set the x limit between jan and june
#plt.xlim("Jan","June")
plt.xlim(0, 11)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()

# Add figure size
plt.figure(figsize=(20,20))
```

```python
plt.barh(x_axis,y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()

plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")

plt.legend()
```

```python
# Create chart using OO functions
fig, ax = plt.subplots(figsize=(20,10))
ax.bar(x_axis,y_axis,label='New York',color="k")

ax.set_ylim(0,45)
ax.set_xlim("Jan","Dec")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

ax.set_title("PyBer Dare by Month")

ax.legend()

#ax.grid()
```
