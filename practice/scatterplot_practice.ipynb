```python
#import dependecies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Using  plt.plot()
#plt.plot(x_axis,y_axis,"o")
```

```python
# Using plot.scatter() add label and color - 5.1.7 Skill Drill
plt.scatter(x_axis,y_axis,c='red',label='Chicago')

# add x and y axis labels
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# add chart title
plt.title("PyBer Fare by Month")
# add legend to the chart
plt.legend()

# Switch the axis so that fare is x -axis; y axis is in asscending order with gca
plt.gca().invert_yaxis()

# add limit to y-axis
plt.ylim(0,50)

plt.show()
```

```python
#plt.scatter(x_axis,y_axis, s=y_axis)

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# Bigger dot size using list comprehension
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis],c='red')
```

```python
#Scatter plot using OO Interface
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# OO Interface - Skill Drill
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis,s=[i*5 for i in y_axis],c='skyblue',alpha=0.2,edgecolor='black',linewidth=2,label='Boston')
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,50)
#plt.gca().invert_yaxis()

plt.show()
```
